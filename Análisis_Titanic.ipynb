 Importación de Librerías

```python
#Trabajamos con Arreglos
import numpy as np
#Trabajar con DataFrames
import pandas as pd
#2 Librerías para Gráficas
import matplotlib.pyplot as plt
import seaborn as sns
#Ignorar warnings
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline 
```

Importar el archivo ".csv" correspondiente

```python
data = pd.read_csv("train_titanic.csv")
```

Reconocemos que columnas tienen nulidad en sus variables

```python
data.isnull().sum()
#Existen valores nulos en Age, Cabin, Embarked
```

Hemos Obtenido nuestra Variable Objetivo, Ahora le haremos una gráfica. 

```python
f, ax = plt.subplots(1,2,figsize = (18,8))
data["Survived"].value_counts().plot.pie(autopct = "%1.1f%%", ax = ax[0], explode = [0,0.1], shadow = True)
ax[0].set_title("Cantidad de Sobrevivientes General")
ax[0].set_ylabel("")


sns.countplot("Sex",hue = "Survived",data = data, ax = ax[1])
ax[1].set_title("Cantidad de Sobrevivientes por Sexo")
f.show()
```

Resultados de la Gráfica Simple
* Las Personas que no Sobrevivieron representan un 61.6 % del total, comprobándose la información exploratoria inicial
* Las Personas que Sobrevivieron son menos de la mitad, representan un 38.4% <br>
-- Sin embargo Nuestro Asunto podría ser mejor apreciado si hacemos una gráfica de este tema teniendo en cuenta el sexo del sobreviviente. Donde:
* Las mujeres presentan una mejor tasa de superviviencia con respecto a los hombres.
* La Diferencia entre sobrevivientes y no Sobrevivientes de los Hombres es alarmante: aprox(100) vs +(400)
